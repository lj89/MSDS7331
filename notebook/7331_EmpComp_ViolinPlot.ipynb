### Violin plot of the continuous variables
These violin plots show the distribution of each of the continuous variables. Each of the following plots only have 3 attributes apiece to allow for the curvature to be seen better.

It is interesting to note that on all attributes except 'Other Salaries' and 'Overtime' there are two noticable peaks.

```python
# Figure size in inches
rcParams['figure.figsize'] = 15, 8
lst = ['Salaries', 'Total Salary', 'Total Compensation']
sns.violinplot(data=df_clean[lst], width = 1)
```

```python
lst = ['Other Salaries', 'Overtime', 'Total Benefits']
sns.violinplot(data=df_clean[lst], width = 1.2)
```
