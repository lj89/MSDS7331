```python
#Packages
import numpy as np  
import matplotlib.pyplot as plt  
get_ipython().run_line_magic('matplotlib', 'inline')
import pandas as pd  
from apyori import apriori
```

```python
#Load data
df_mlt = pd.read_csv('/Users/mtran/Documents/SMUDataSciences/2019-Summer/Diabetic_Cleaned.csv')
print("Finished loading data")
```

```python
#Make dataframes for each column
ordinal = ["age"]
categorical = ["gender", "max_glu_serum", "A1Cresult","diabetesMed"]
medicines = ["insulin"]
```

```python
df_mlt_ordinal = df_mlt[ordinal]
df_mlt_categorical = df_mlt[categorical]
df_mlt_medicines = df_mlt[medicines]
```

```python
#Make age ordinal
for i in range(0, 10):
    val = "[" + str(i*10) + "-" + str((i+1)*10) + ")"
    df_mlt_ordinal.loc[df_mlt_ordinal['age'] == val, 'age'] = i
```

```python
#Differentiate A1Cresult values from max_glu_serum
df_mlt_categorical.A1Cresult[df_mlt_categorical.A1Cresult == 'None'] = 'A1C None'
df_mlt_categorical.A1Cresult[df_mlt_categorical.A1Cresult == 'Norm'] = 'A1C Norm'
df_mlt_categorical.A1Cresult[df_mlt_categorical.A1Cresult == '>7'] = 'A1C >7'
df_mlt_categorical.A1Cresult[df_mlt_categorical.A1Cresult == '>8'] = 'A1C >8'
df_mlt_categorical.diabetesMed[df_mlt_categorical.diabetesMed == 'No'] = 'DM No'
df_mlt_categorical.diabetesMed[df_mlt_categorical.diabetesMed == 'Yes'] = 'DM Yes'
```

#### Creating dataframe

```python
#Creating dataframe
df_mlt_med_x = pd.concat([df_mlt_ordinal, df_mlt_categorical, df_mlt_medicines], axis = 1)
print("Done")
```

```python
#Print dataframe
df_mlt_med_x
```

```python
num_records = len(df_mlt_med_x)
print(num_records)
```

```python
%%time

records = []  
for i in range(0, num_records):  
    records.append([str(df_mlt_med_x.values[i,j]) for j in range(0, 6)]) 
```

```python
%%time

association_rules = apriori(records, min_support=0.01, min_confidence=0.5, min_lift=1.1, min_length=2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

#### Creating dataframe of rules 

```python
%%time

rules_df = pd.DataFrame(columns=['Rule', 'Support', 'Confidence', 'Lift'])

for item in association_results:
    pair = item[0] 
    items = [x for x in pair]
    rl = str(items[0] + " -> " + items[1])
    sp = str(item[1])
    cf = str(item[2][0][2])
    lf = str(item[2][0][3])
    
    rules_df.loc[len(rules_df), :] = [rl, sp, cf, lf]
```

#### Print and sort rules by lift, support, and confidence

```python
#Sort rules descending by lift
rules_df.sort_values(by='Lift', ascending=False)
```

```python
#Sort rules descending by Support
rules_df.sort_values(by='Support', ascending=False)
```

```python
#Sort rules descending by confidence
rules_df.sort_values(by='Confidence', ascending=False)
```
