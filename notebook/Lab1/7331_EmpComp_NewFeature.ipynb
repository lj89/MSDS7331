### New Feature
Based on the density and violin plots from the Simple Statistics portion of this homework, it appears that the data has (at least) two local maximums. Because of this we have considered that our target regression attribute 'Total Salaries' could work well as a classification variable. By this we mean that we would make a new column called 'Total Salary Group' which would be comprised of 'Low', 'Medium', and 'High'. Below are some plots that demonstrate this effect on 'Job Family' vs 'Total Salary' and 'Organization Group' vs 'Total Salary'.

```python
sns.FacetGrid(df_clean, hue="Job Family", size=6) \
.map(sns.kdeplot, "Total Salary") \
.add_legend()
```

```python
sns.FacetGrid(df_clean, hue="Organization Group", size=6) \
.map(sns.kdeplot, "Total Salary") \
.add_legend()
```
